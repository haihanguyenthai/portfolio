# **Feature Enginnering: Credibility Definition**

1. Điều kiện cho Credible Debt (Nợ đáng tin cậy):
Nhóm nợ: Khách hàng thuộc nhóm nợ 1 được coi là đáng tin cậy. Đây là nhóm khách hàng có lịch sử tín dụng tốt, ít có nguy cơ vỡ nợ. Theo các nghiên cứu trong ngành ngân hàng, khách hàng có nhóm nợ thấp thường có khả năng thanh toán tốt hơn (Jansson et al., 2020).

Số dư tài khoản: Số dư tài khoản được so sánh với percentile 25. Nếu số dư tài khoản của khách hàng nằm ở mức cao hơn 25% so với những khách hàng khác, có thể cho thấy khách hàng có tài chính ổn định. Các ngân hàng thường đánh giá khả năng thanh toán của khách hàng dựa trên số dư tài khoản hiện tại (Luo et al., 2021).

Lãi suất thực tế: Lãi suất thực tế thấp (<= 25%) cho thấy khách hàng có thể dễ dàng thanh toán nợ mà không bị quá tải tài chính. Theo các nghiên cứu về phân tích rủi ro tín dụng, lãi suất thực tế thấp giúp khách hàng có thể trả nợ dễ dàng hơn mà không gặp phải khó khăn tài chính nghiêm trọng (Lee et al., 2020).

Tổng giao dịch tháng: Điều kiện này yêu cầu tổng giao dịch tháng của khách hàng phải ở mức trung vị. Điều này nhằm đánh giá mức độ ổn định tài chính của khách hàng thông qua các giao dịch tài chính hàng tháng. Khách hàng có giao dịch đều đặn thường ít có khả năng vỡ nợ hơn (Shao et al., 2021).

Loại kỳ hạn khoản vay: Chỉ các khoản vay có kỳ hạn dài (DH) hoặc trung (TH) mới được xem xét. Kỳ hạn dài hoặc trung bình giúp giảm áp lực trả nợ trong ngắn hạn, làm giảm nguy cơ vỡ nợ (Shao et al., 2021).

2. Điều kiện cho Incredible Debt (Nợ không đáng tin cậy):
Nhóm nợ: Khách hàng thuộc nhóm nợ từ 2 trở lên có thể bị coi là không đáng tin cậy, vì nhóm này có nguy cơ vỡ nợ cao hơn. Ngành ngân hàng sử dụng nhóm nợ như một yếu tố quan trọng để xác định khách hàng có nguy cơ vỡ nợ hay không (Luo et al., 2021).

Lãi suất thực tế và danh nghĩa: Khách hàng có lãi suất thực tế cao hoặc lãi suất danh nghĩa lớn hơn một mức cố định cho thấy khách hàng có thể đang gặp khó khăn tài chính và sẽ khó thanh toán nợ. Các nghiên cứu chỉ ra rằng lãi suất cao thường đồng nghĩa với việc khách hàng phải đối mặt với các khoản vay không trả được hoặc các khoản phí cao hơn (Lee et al., 2020).

Giao dịch tín dụng: Điều kiện này kiểm tra xem khách hàng có giao dịch tín dụng thường xuyên hay không. Các khách hàng có giao dịch tín dụng nhiều có thể gặp phải tình trạng nợ xấu hoặc vỡ nợ (Jansson et al., 2020).

Giao dịch tháng cao: Các khách hàng có giao dịch tài chính cao trong các tháng gần đây (tháng 3 và tháng 6) có thể là dấu hiệu của sự căng thẳng tài chính, khi họ phải huy động tiền để thanh toán các khoản nợ trước đó.

Giao dịch lớn nhất tuần: Điều kiện này yêu cầu khách hàng có giao dịch lớn nhất tuần cao hơn mức trung vị. Điều này có thể chỉ ra khách hàng đang sử dụng nhiều tín dụng, và khi giao dịch lớn xảy ra thường xuyên, điều này có thể làm tăng khả năng vỡ nợ (Shao et al., 2021).


Dùng điều kiện để tìm credible customer và incredible customer

```python

df['debt_evaluation'] = -1  

# Điều kiện cho Credible Debt (Nợ đáng tin cậy)
credible_conditions = (
    (df['nhomno_xhtdJuin2021'] == 1) &  # Nhóm nợ thuộc nhóm 1 (đáng tin cậy)
    (df['balanceJuin2021'] >= df['balanceJuin2021'].quantile(0.25)) &  # Số dư tài khoản >= 25th percentile
    (df['real_interestJuin2021'] <= 0.25) &  # Lãi suất thực tế ≤ 25%
    (df['amount_month'] >= df['amount_month'].quantile(0.50)) &  # Tổng giao dịch tháng >= median
    (df['loaikyhanJuin2021'].isin(['DH', 'TH']))  # Chỉ tính kỳ hạn dài (DH) và trung (TH)
)


# Đánh dấu các khách hàng là Credible Debt (0)
df.loc[credible_conditions, 'debt_evaluation'] = 0


incredible_conditions = (
    (df['nhomno_xhtdJuin2021'] >= 2) &  # Nhóm nợ từ 2 trở lên
    (
        (df['real_interestJuin2021'] > 0.0) |  # Lãi suất thực tế cao hơn trung vị
        (df['nominal_interestJuin2021'] > 0.0001)  # Lãi suất danh nghĩa cao hơn trung vị
    ) &
    (
        (df['distinct_payment_code_week'] >= 1) |  # Có giao dịch tín dụng
        (df['count_payment_code_week'] > 1)  # Số lượng giao dịch tín dụng > 1
    ) &
    (
        (df['most_act_mar2021_count'] > 8) |  # Giao dịch tháng 3 cao
        (df['most_act_juin2021_count'] > 8)  # Giao dịch tháng 6 cao
    ) &
    (df['max_amount_week'] > 0.00015)  # Giao dịch lớn nhất tuần > median
)

# Đánh dấu các khách hàng là Incredible Debt (1)
df.loc[incredible_conditions, 'debt_evaluation'] = 1

# Kiểm tra kết quả phân loại
print(df['debt_evaluation'].value_counts())  # Xem số lượng các loại nợ (0 - Credible, 1 - Incredible)
```

```python
import pandas as pd
from kmodes.kmodes import KModes
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# Step 1: Load your data
# Assuming `df` is the DataFrame containing your dataset
data = df.copy()

# Step 2: Replace NaN with an empty string specifically for categorical columns
categorical_columns = data.select_dtypes(include=['object', 'category']).columns
data[categorical_columns] = data[categorical_columns].fillna('')

# Step 3: Convert all categorical columns to strings
for col in categorical_columns:
    data[col] = data[col].astype(str)

# Step 4: Select only categorical columns for clustering
features = data[categorical_columns]

# Step 5: Perform K-modes clustering (only 2 clusters)
k_modes = KModes(n_clusters=2, init='Huang', n_init=5, random_state=42)
clusters = k_modes.fit_predict(features)

# Assign clusters to the original DataFrame
data['cluster'] = clusters

# Step 6: Filter records where `debt_evaluation` == -1
df_negative = data[data['debt_evaluation'] == -1]

# Step 7: Use PCA to reduce dimensionality for visualization (since K-modes works on categorical data)
# PCA requires numerical data, so we first need to encode the categorical data (one-hot encoding)
encoded_data = pd.get_dummies(df_negative[categorical_columns])

# Apply PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(encoded_data)

# Create a DataFrame for PCA results
df_pca = pd.DataFrame(pca_result, columns=['PCA1', 'PCA2'])
df_pca['cluster'] = df_negative['cluster'].values

# Step 8: Plot the clusters in a 2D space
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_pca, x='PCA1', y='PCA2', hue='cluster', palette='Set2', s=100)

# Add labels and title
plt.title('K-modes Clustering with 2 Clusters for debt_evaluation = -1', fontsize=14)
plt.xlabel('PCA1', fontsize=12)
plt.ylabel('PCA2', fontsize=12)
plt.legend(title='Cluster', fontsize=10)
plt.show()
```

```python


# Separate known and unknown data
known_data = df[df['debt_evaluation'].isin([0, 1])]
unknown_data = df[df['debt_evaluation'] == -1]

# Drop 'debt_evaluation' from features
X_known = known_data.drop(columns=['debt_evaluation'])
y_known = known_data['debt_evaluation']
X_unknown = unknown_data.drop(columns=['debt_evaluation'])

# Step 2: Compute Hamming distance
# Convert all categorical columns to strings for uniformity
X_known = X_known.astype(str)
X_unknown = X_unknown.astype(str)

# Encode each column independently
for col in X_known.columns:
    unique_values = pd.concat([X_known[col], X_unknown[col]]).unique()
    value_map = {val: idx for idx, val in enumerate(unique_values)}
    X_known[col] = X_known[col].map(value_map)
    X_unknown[col] = X_unknown[col].map(value_map)

# Step 3: Compute Hamming distances
# Convert DataFrames to NumPy arrays
X_known_array = X_known.to_numpy()
X_unknown_array = X_unknown.to_numpy()

# Compute Hamming distances between unknown and known rows
hamming_distances = cdist(X_unknown_array, X_known_array, metric='hamming')

# Step 4: Find the nearest neighbors and predict
k = 5  # Number of neighbors to consider
predictions = []

for i, distances in enumerate(hamming_distances):
    # Get the indices of the k nearest neighbors
    nearest_indices = np.argsort(distances)[:k]
    nearest_labels = y_known.iloc[nearest_indices]
    
    # Majority vote
    most_common = Counter(nearest_labels).most_common(1)[0][0]
    predictions.append(most_common)

# Step 5: Replace -1 with predicted values
df.loc[df['debt_evaluation'] == -1, 'debt_evaluation'] = predictions

# Verify the updated dataset
print(df['debt_evaluation'].value_counts())
```

## **Encoding**

**Tóm tắt giải thích:**  

- **Cột `product_codeJuin2021` và `sectorJuin2021`:** Các giá trị xuất hiện dưới 90% được thay thế bằng giá trị phổ biến nhất để giảm chiều dữ liệu và cải thiện hiệu suất. Sau đó, mã hóa bằng **One-Hot Encoding** để biểu diễn mỗi danh mục dưới dạng cột nhị phân, tránh hiểu nhầm quan hệ định lượng giữa các danh mục.  

- **Cột `nhomno_xhtdJuin2021`:** Mã hóa bằng **custom ranking encoding**, ánh xạ các nhóm nợ theo thứ tự rủi ro, với nhóm đáng tin cậy nhất được mã hóa là 0. Phương pháp này giúp mô hình hiểu được ý nghĩa thứ tự giữa các nhóm.  

- **Cột `sex`:** Mã hóa nhị phân, với "FEMALE" là 0 và "MALE" là 1, giúp đơn giản hóa dữ liệu và tăng hiệu quả xử lý.  

- **Các cột danh mục khác:** Sử dụng **One-Hot Encoding** cho các biến như `marital_status`, `categoryJuin2021`, `loaikyhanJuin2021`, `most_act_mar2021`, và `most_act_juin2021`, đảm bảo loại bỏ hiểu nhầm quan hệ thứ tự và chuẩn bị dữ liệu cho mô hình học máy.

```python

# Tính tần suất các giá trị trong cột 'product_codeJuin2021' và 'sectorJuin2021'
product_code_counts = df1['product_codeJuin2021'].value_counts(normalize=True)
sector_counts = df1['sectorJuin2021'].value_counts(normalize=True)

# Lọc ra các giá trị có tần suất >= 90% (high frequency)
high_freq_product_code = product_code_counts[product_code_counts >= 0.9].index
high_freq_sector = sector_counts[sector_counts >= 0.9].index

# Gán giá trị theo tần suất cao nhất cho các category còn lại
# Đối với cột 'product_codeJuin2021'
df1['product_codeJuin2021'] = df1['product_codeJuin2021'].apply(
    lambda x: x if x in high_freq_product_code else product_code_counts.idxmax())

# Đối với cột 'sectorJuin2021'
df1['sectorJuin2021'] = df1['sectorJuin2021'].apply(
    lambda x: x if x in high_freq_sector else sector_counts.idxmax())

# Tạo dummies cho các cột sau khi đã thay giá trị cho các category còn lại
df1 = pd.get_dummies(df1, columns=['product_codeJuin2021', 'sectorJuin2021'], drop_first=False)
```

```python
# 1. Special encoding for 'nhomno_xhtdJuin2021' with custom ranking (descending order)
unique_values = sorted(df1['nhomno_xhtdJuin2021'].unique(), reverse=True)
custom_mapping = {value: rank for rank, value in enumerate(unique_values)}

df1['nhomno_xhtdJuin2021_encoded'] = df1['nhomno_xhtdJuin2021'].map(custom_mapping)

# 2. Encode 'sex' with a predefined mapping
sex_encoding = {"FEMALE": 0, "MALE": 1}
df1['sex'] = df1['sex'].replace(sex_encoding)

# 4. Normalize category columns ('most_act_mar2021' and 'most_act_juin2021')
# Convert the values to lowercase to ensure uniformity for further encoding
df1['most_act_mar2021'] = df1['most_act_mar2021'].str.lower()
df1['most_act_juin2021'] = df1['most_act_juin2021'].str.lower()

# 5. Apply One-Hot Encoding for categorical columns
df1 = pd.get_dummies(df1, columns=['marital_status', 'categoryJuin2021', 'loaikyhanJuin2021', 
                                   'most_act_mar2021', 'most_act_juin2021'], drop_first=False)

# Displaying the updated dataframe (optional)
df1.head()
```

**Learning Curve**

```python
def plot_learning_curve(estimator, X, y, cv, train_sizes):
    train_sizes, train_scores, test_scores = learning_curve(estimator, X, y, cv=cv, train_sizes=train_sizes)

    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    plt.figure(figsize=(8, 6))
    plt.plot(train_sizes, train_mean, 'o-', color='r', label='Training Score')
    plt.plot(train_sizes, test_mean, 'o-', color='b', label='Cross-Validation Score')
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='r')
    plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.1, color='b')
    plt.xlabel('Training Set Size')
    plt.ylabel('Score')
    plt.title('Learning Curve')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()
```
